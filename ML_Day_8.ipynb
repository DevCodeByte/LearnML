```python
import matplotlib.pyplot as plt
import numpy as np
# np.random.normal(170, 10, 250)
# This line is where the magic happens. Let's decode it:
# 170: This is the mean (average) of the distribution you want. Your randomly generated numbers will tend to cluster around 170.
# 10: This is the standard deviation. It controls the spread or width of the bell curve. A larger standard deviation means the numbers will be more spread out.
# 250: This is the size of the sample. It tells the function to generate 250 random numbers.

x = np.random.normal(170,10,250)
print(x)
plt.hist(x)
```
