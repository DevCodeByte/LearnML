```python
df = pd.read_csv('dirtydata.csv')
print(df)
```

```python
print(df.info())
print(df.describe()) #It will not display date which a object
```

```python
#cleaning data
new_df = df.dropna()
print(new_df.info())
```

```python
#find the mean in a column
x = df["Calories"].mean()
print(x)
```

```python
#fill the nan with mean value
newdf = df.fillna(x)
print(newdf)
#only replace in the respective column
#df["Calories"].fillna(x, inplace = True)
#OR
#df.fillna({'Calories':x}, inplace=True)
#OR
df["Calories"] = df["Calories"].fillna(x)
print(df)
```

```python
# lets replace nan with median
y = df["Calories"].median()
print(y)
print(type(y))

#lets replace nan with mod
zzz = df["Calories"].mode()
print(zzz)
print(type(zzz))

#As mode is type of Series i.e. <class 'pandas.core.series.Series'>, we need to get index
z = df["Calories"].mode()[0] # same as zzz[0]
print(z)
print(type(z))
```
